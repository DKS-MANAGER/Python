## Day 1 — Classes

Usefull OOP stuff — class is like a blueprint , object is the actual thing
same idea as the `Car` class but now with real examples from college life

````python
# class = blueprint , object = real thing
# like Car class i made in notebook but now actually useful

class Student:
    college = "IIT Guwahati"   # class variable - same for all students

    def __init__(self, name, roll, branch, gpa):
        self.name   = name
        self.roll   = roll
        self.branch = branch
        self.gpa    = gpa

    def greet(self):
        print(f"Hi i'm {self.name} | Roll: {self.roll} | {self.branch} @ {Student.college}")

    def is_topper(self):
        return self.gpa >= 9.0

    def cgpa_status(self):
        if self.gpa >= 9.0:
            return "Outstanding"
        elif self.gpa >= 8.0:
            return "Excellent"
        elif self.gpa >= 6.0:
            return "Good"
        else:
            return "Needs improvement"

# creating objects - my batchmates
s1 = Student("divyansh",  "22110045", "Civil", 8.5)
s2 = Student("rahul",     "22110032", "Mech",  9.2)
s3 = Student("sneha",     "22110078", "CSE",   7.8)

s1.greet()
s2.greet()
s3.greet()

print("\ncgpa status:")
for s in [s1, s2, s3]:
    print(f"  {s.name:10s} gpa={s.gpa}  → {s.cgpa_status()}")

print(f"\ntopper check: {s2.name} is topper = {s2.is_topper()}")
print("num_students:", Student.college)
````

````python
# Expense class - core of the week2 project
# tracking my hostel + college expenses

class Expense:
    def __init__(self, date, category, amount, note=""):
        self.date     = date
        self.category = category
        self.amount   = float(amount)
        self.note     = note

    def summary(self):
        return f"[{self.date}]  {self.category:12s}  ₹{self.amount:.2f}  ({self.note})"

# my march expenses
e1 = Expense("2026-03-01", "Food",      250,  "canteen lunch")
e2 = Expense("2026-03-01", "Transport", 80,   "auto to college")
e3 = Expense("2026-03-02", "Books",     450,  "Fluid Mechanics book")
e4 = Expense("2026-03-02", "Food",      180,  "chai + maggi")
e5 = Expense("2026-03-02", "Internet",  399,  "Jio recharge")

expenses = [e1, e2, e3, e4, e5]

print("Expense summary:")
for e in expenses:
    print(" ", e.summary())

total = sum(e.amount for e in expenses)
print(f"\nTotal spent : ₹{total:.2f}")
````

## Day 2 — Inheritance & Polymorphism

`__repr__` and `__str__` — without these you just get `<object at 0x...>` which is useless for debugging
inheritance = child class gets all parent stuff + can add its own

````python
# __str__ = what user sees (print)
# __repr__ = what dev sees (repr / debugger)

class Expense:
    def __init__(self, date, category, amount, note=""):
        self.date     = date
        self.category = category
        self.amount   = float(amount)
        self.note     = note

    def __str__(self):
        return f"[{self.date}] {self.category} ₹{self.amount:.2f} — {self.note}"

    def __repr__(self):
        return f"Expense(date='{self.date}', cat='{self.category}', amt={self.amount})"

    def __eq__(self, other):
        # same amount + same category = equal (note doesnt matter)
        return self.amount == other.amount and self.category == other.category

e1 = Expense("2026-03-01", "Food", 250, "canteen")
e2 = Expense("2026-03-02", "Food", 250, "dinner")
e3 = Expense("2026-03-01", "Books", 450, "textbook")

print(e1)           # __str__
print(repr(e1))     # __repr__
print(f"e1 == e2 : {e1 == e2}")   # True - same cat + amount
print(f"e1 == e3 : {e1 == e3}")   # False
````

````python
# inheritance - same as Animal/Dog/Cat i did in notebook
# but now base = Person , children = Student , Professor

class Person:
    def __init__(self, name, age):
        self.name     = name
        self.age      = age
        self.is_alive = True

    def introduce(self):
        print(f"Hi i'm {self.name} , age {self.age}")

    def __repr__(self):
        return f"Person({self.name})"

class Student(Person):
    def __init__(self, name, age, roll, branch):
        super().__init__(name, age)   # call parent first
        self.roll   = roll
        self.branch = branch

    def introduce(self):              # override parent version
        print(f"Hi i'm {self.name} | Roll {self.roll} | {self.branch} @ IITG")

    def __repr__(self):
        return f"Student({self.name}, {self.roll})"

class Professor(Person):
    def __init__(self, name, age, subject):
        super().__init__(name, age)
        self.subject = subject

    def introduce(self):
        print(f"I'm Prof. {self.name} , i teach {self.subject}")

    def __repr__(self):
        return f"Professor({self.name})"

# polymorphism - same method name , python chooses right version
people = [
    Student("divyansh", 21, "22110045", "Civil"),
    Student("rahul",    22, "22110032", "Mech"),
    Professor("Dr. Sharma", 45, "Fluid Mechanics"),
    Professor("Dr. Roy",    50, "Engineering Maths"),
]

print("Introductions:")
for p in people:
    p.introduce()
    print(f"  repr → {repr(p)}")
````

## Day 3 — File I/O , CSV , JSON

Usefull for saving data permanently — variables disappear when kernel restarts , files dont
always use `with` — it closes the file automatically even if an error happens

````python
# text file - write / read / append
# 'with' = context manager , auto closes file

# writing my week notes
with open("my_notes.txt", "w") as f:
    f.write("Week 2 notes — divyansh Kumar sinGH\n")
    f.write("Roll: 22110045 | Branch: Civil | IITG\n")
    f.write("Topics: OOP , File I/O , Exceptions , Testing\n")
    f.write("date: 2026-03-02\n")

print("written ✓")

# reading full file at once
with open("my_notes.txt", "r") as f:
    content = f.read()
print(content)

# reading line by line
with open("my_notes.txt", "r") as f:
    for i, line in enumerate(f, 1):
        print(f"  line {i}: {line.strip()}")

# appending without overwriting
with open("my_notes.txt", "a") as f:
    f.write("progress: day 3 done\n")

print("appended ✓")
````

````python
import csv

# writing to CSV — like excel but plain text
# my march expenses
expenses_data = [
    ["date",       "category",  "amount", "note"],
    ["2026-03-01", "Food",      250,      "canteen lunch"],
    ["2026-03-01", "Transport", 80,       "auto to college"],
    ["2026-03-02", "Books",     450,      "Fluid Mechanics book"],
    ["2026-03-02", "Food",      180,      "chai + maggi"],
    ["2026-03-02", "Internet",  399,      "Jio recharge"],
    ["2026-03-02", "Clothing",  650,      "new hoodie"],
]

with open("expenses.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(expenses_data)

print("expenses.csv created ✓")

# reading back with DictReader — gives dict per row
category_total = {}

with open("expenses.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        cat    = row["category"]
        amount = float(row["amount"])
        category_total[cat] = category_total.get(cat, 0) + amount

print("\nSpending by category:")
for cat, total in sorted(category_total.items(), key=lambda x: -x[1]):
    print(f"  {cat:12s}  ₹{total:.2f}")

print(f"\nTotal : ₹{sum(category_total.values()):.2f}")
````

## Day 4 — Exceptions , Context Managers & Logging

custom exceptions = way better than crashing with a random error
`with` = context manager — opens + closes resource automatically
`logging` = better than `print` for real projects (has levels , timestamps)

````python
# custom exceptions - much cleaner than letting it crash

class InvalidAmountError(Exception):
    pass   # amount <= 0

class InvalidCategoryError(Exception):
    pass   # category not in allowed list

ALLOWED_CATEGORIES = ["Food", "Transport", "Books", "Internet", "Clothing", "Other"]

def add_expense(category, amount, note=""):
    if amount <= 0:
        raise InvalidAmountError(f"amount must be positive , got {amount}")
    if category not in ALLOWED_CATEGORIES:
        raise InvalidCategoryError(f"'{category}' not valid — pick from {ALLOWED_CATEGORIES}")
    return {"category": category, "amount": amount, "note": note}

# good input
try:
    e = add_expense("Food", 250, "canteen")
    print("Added :", e)
except (InvalidAmountError, InvalidCategoryError) as err:
    print("Error :", err)

# bad amount
try:
    add_expense("Food", -50)
except InvalidAmountError as err:
    print("Caught :", err)

# bad category - i always forget the exact name
try:
    add_expense("Chai", 30)
except InvalidCategoryError as err:
    print("Caught :", err)

# finally - runs no matter what
try:
    x = int("not a number")
except ValueError as err:
    print("ValueError :", err)
finally:
    print("finally block always runs")
````

````python
import logging

# logging levels (low to high) :
# DEBUG < INFO < WARNING < ERROR < CRITICAL
# set level = show that level and above only

logging.basicConfig(
    level   = logging.DEBUG,
    format  = "%(asctime)s  %(levelname)-8s  %(message)s",
    datefmt = "%H:%M:%S"
)

log = logging.getLogger("tracker")

log.info("Expense Tracker started — divyansh 22110045")
log.debug("loading expenses.csv ...")

import csv, os

def load_expenses(filepath):
    if not os.path.exists(filepath):
        log.error(f"file not found : {filepath}")
        return []
    with open(filepath) as f:
        rows = list(csv.DictReader(f))
    log.info(f"loaded {len(rows)} rows from {filepath}")
    return rows

rows = load_expenses("expenses.csv")

for r in rows:
    log.debug(f"  {r['date']}  {r['category']:12s}  ₹{float(r['amount']):.2f}")

total = sum(float(r["amount"]) for r in rows)
if total > 1500:
    log.warning(f"budget getting high ! total = ₹{total:.2f}")

log.info("session ended")
````

````python
# tracker.py — CLI Expense Tracker
# divyansh Kumar sinGH | 22110045 | week2 deliverable

import csv
import os
from datetime import date

FILENAME = "expenses.csv"
CATEGORIES = ["Food", "Transport", "Books", "Internet", "Clothing", "Other"]

def init_file():
    """make csv with header if it doesnt exist"""
    if not os.path.exists(FILENAME):
        with open(FILENAME, "w", newline="") as f:
            csv.writer(f).writerow(["date", "category", "amount", "note"])
        print(f"{FILENAME} created")

def add_expense(category, amount, note=""):
    """add one expense"""
    if amount <= 0:
        raise ValueError(f"amount must be positive , got {amount}")
    if category not in CATEGORIES:
        raise ValueError(f"invalid category : {category}")
    with open(FILENAME, "a", newline="") as f:
        csv.writer(f).writerow([date.today(), category, round(amount, 2), note])

def get_all():
    """return all expenses as list of dicts"""
    if not os.path.exists(FILENAME):
        return []
    with open(FILENAME) as f:
        return list(csv.DictReader(f))

def total_by_category():
    """category → total ₹"""
    result = {}
    for row in get_all():
        cat = row["category"]
        result[cat] = result.get(cat, 0) + float(row["amount"])
    return result

def monthly_total():
    """grand total"""
    return sum(float(r["amount"]) for r in get_all())

def show_all():
    rows = get_all()
    if not rows:
        print("no expenses yet")
        return
    print(f"\n{'Date':12} {'Category':12} {'Amount':>8}  Note")
    print("-" * 50)
    for r in rows:
        print(f"  {r['date']:12} {r['category']:12} ₹{float(r['amount']):>6.2f}  {r['note']}")
    print("-" * 50)
    print(f"  {'TOTAL':25} ₹{monthly_total():>6.2f}")

# ── demo run ──────────────────────────────────────────────
os.remove(FILENAME) if os.path.exists(FILENAME) else None   # fresh start

init_file()
add_expense("Food",      250,  "canteen lunch")
add_expense("Transport", 80,   "auto to college")
add_expense("Books",     450,  "Fluid Mech book")
add_expense("Food",      180,  "chai + maggi")
add_expense("Internet",  399,  "Jio recharge")
add_expense("Clothing",  650,  "new hoodie flipkart")

show_all()

print("\nBy category:")
for cat, tot in sorted(total_by_category().items(), key=lambda x: -x[1]):
    print(f"  {cat:12s}  ₹{tot:.2f}")
````

## Day 6 — Unit Tests (`unittest` / `pytest`)

testing = making sure your code doesnt break when you change something
write tests once , run them every time you edit the code

**two options :**
- `unittest` — built into python , no install needed
- `pytest` — cleaner syntax , better output , industry standard

````python
import unittest
import os
import csv
from datetime import date

# ── test file for tracker functions ───────────────────────────────────────────
# in real project this goes in tests/test_tracker.py
# here i just define + run it inline

TEST_FILE = "test_expenses.csv"

# patch the tracker to use a test file so we don't mess up real data
def init_test_file():
    with open(TEST_FILE, "w", newline="") as f:
        csv.writer(f).writerow(["date", "category", "amount", "note"])

def add_test_expense(category, amount, note=""):
    CATEGORIES = ["Food", "Transport", "Books", "Internet", "Clothing", "Other"]
    if amount <= 0:
        raise ValueError(f"amount must be positive , got {amount}")
    if category not in CATEGORIES:
        raise ValueError(f"invalid category : {category}")
    with open(TEST_FILE, "a", newline="") as f:
        csv.writer(f).writerow([date.today(), category, round(amount, 2), note])

def get_test_all():
    with open(TEST_FILE) as f:
        return list(csv.DictReader(f))

def test_monthly_total():
    return sum(float(r["amount"]) for r in get_test_all())


class TestTracker(unittest.TestCase):

    def setUp(self):
        """runs before each test - fresh csv each time"""
        init_test_file()

    def tearDown(self):
        """runs after each test - cleanup"""
        if os.path.exists(TEST_FILE):
            os.remove(TEST_FILE)

    def test_add_expense_normal(self):
        add_test_expense("Food", 250, "canteen")
        rows = get_test_all()
        self.assertEqual(len(rows), 1)
        self.assertEqual(rows[0]["category"], "Food")
        self.assertEqual(float(rows[0]["amount"]), 250.0)

    def test_bad_amount(self):
        with self.assertRaises(ValueError):
            add_test_expense("Food", -50)

    def test_zero_amount(self):
        with self.assertRaises(ValueError):
            add_test_expense("Food", 0)

    def test_invalid_category(self):
        with self.assertRaises(ValueError):
            add_test_expense("Chai", 30)   # not in allowed list

    def test_monthly_total(self):
        add_test_expense("Food",      250, "canteen")
        add_test_expense("Transport", 80,  "auto")
        add_test_expense("Books",     450, "textbook")
        self.assertAlmostEqual(test_monthly_total(), 780.0)

    def test_multiple_entries(self):
        add_test_expense("Food", 100)
        add_test_expense("Food", 150)
        add_test_expense("Books", 300)
        rows = get_test_all()
        self.assertEqual(len(rows), 3)


# run tests inline in notebook
loader = unittest.TestLoader()
suite  = loader.loadTestsFromTestCase(TestTracker)
runner = unittest.TextTestRunner(verbosity=2)
result = runner.run(suite)

print(f"\nTests run   : {result.testsRun}")
print(f"Failures    : {len(result.failures)}")
print(f"Errors      : {len(result.errors)}")
print("All passed  ✓" if result.wasSuccessful() else "some tests FAILED")
````
